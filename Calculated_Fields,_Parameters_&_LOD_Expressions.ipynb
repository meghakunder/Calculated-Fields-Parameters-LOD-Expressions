**Question 1 : What is a Calculated Field and why is it important in business dashboards?**

**ANS:**

A calculated field is a user-defined formula in Tableau used to create new data from existing fields. It is important because it helps generate key metrics and insights, making dashboards more informative and useful for decision-making.

Example: Profit Ratio = Profit / Sales

**Question 2 :Why can’t parameters change visuals directly in Tableau?**

**ANS:**

In Tableau, parameters cannot change visuals directly because they are independent variables that are not automatically linked to the dataset or visualizations.

Parameters only store a single value and do not have any inherent connection to charts, filters, or fields. Therefore, changing a parameter does not affect the visualization unless it is explicitly used within a calculated field, filter, or reference line.

To make parameters interactive, they must be connected to the dashboard through calculations or logic, which then dynamically updates the visual based on the parameter value.

**Question 3 : When should EXCLUDE LOD be used?**

**ANS:**

In Tableau, EXCLUDE LOD (Level of Detail) is used when we want to perform calculations by removing one or more dimensions from the current view’s level of detail.

It is useful when the visualization includes detailed dimensions, but we want to calculate values at a higher or aggregated level. For example, calculating total sales excluding a specific dimension like category or region while still displaying those dimensions in the view.

**Question 4 : Difference between FIXED and INCLUDE LOD expressions?**

**ANS:**

In Tableau, FIXED and INCLUDE are Level of Detail (LOD) expressions used to control the level at which calculations are performed, but they behave differently.

FIXED LOD computes values at a specified level of detail independent of the view. It ignores the dimensions present in the visualization and calculates based only on the dimensions defined in the expression. It is useful when we want a consistent value regardless of what is displayed.

INCLUDE LOD, on the other hand, adds additional dimensions to the current view’s level of detail. It performs calculations at a more granular level and then aggregates the results back to the view level.

In summary, FIXED overrides the view’s level of detail, while INCLUDE extends it.

FIXED:

Total Sales per Region (same value even if Category is added to view)

INCLUDE:

Sales per Customer (even if Customer not in view)

**Task 4: City-Level Benchmarking (FIXED LOD)**

**Business Need**

**Compare city performance independent of category filters.**

**Task**

**Calculate average sales per city using FIXED LOD**

**ANS:**

image.png

a) Calculated Field

A FIXED LOD expression Avg Sales per City (LOD) was created to calculate average sales for each city.

b) Logic

{ FIXED City : AVG(Sales) }
This ensures the calculation is fixed at the city level and remains independent of other filters like Category.

c) Insight

The average sales per city remain constant even when different category filters are applied, providing a stable benchmark for comparison.

d) Business Value

This helps management evaluate true city performance without being influenced by product categories, enabling better strategic decisions related to expansion and operations.

**Task 5: Customer/Product Detail Impact (INCLUDE LOD)**

**Business Need**

**The sales team wants average sales considering product-level detail, even when product is not shown.**

**Task**

**Create a calculation that finds the average sales at the product level, but allows the result to be used at a higher level (such as Category or Region). Use INCLUDE LOD.**

**ANS:**

image.png

a) Calculated Field

An INCLUDE LOD expression was created to calculate average sales at the product level.

b) Logic

{ INCLUDE Product Name : AVG(Sales) }
This adds product-level detail into the calculation even when it is not present in the visualization.

c) Insight

The average sales values reflect product-level performance aggregated at the category level, providing a more granular and accurate view.

d) Business Value

This helps the sales team understand how individual products contribute to overall category performance, enabling better decisions in product prioritization and pricing strategies.
